```python
# dict(list(df.groupby('Country')))['Belgium']
countryfre=[]
# df[df['Country']=='USA']['Country'].count()
countryfre.append(df[df['Country']=='USA']['Country'].count())
countryfre.append(df[df['Country']=='UK']['Country'].count())
countryfre.append(df[df['Country']=='Belgium']['Country'].count())
countryfre.append(df[df['Country']=='Russia']['Country'].count())
countryfre.append(df[df['Country']=='Denmark']['Country'].count())
countryfre.append(df[df['Country']=='Germany']['Country'].count())
countryfre.append(df[df['Country']=='Canada']['Country'].count())
countryfre.append(df[df['Country']=='Mexico']['Country'].count())

countryfre.sort(reverse=True)
countryfre
ind=df['Country'].unique()
len(ind)
ind=ind[0:-1]
final_df=pd.DataFrame(countryfre,index=ind,columns=['freq'])
final_df['Relative']=final_df['freq']/final_df['freq'].sum()*100
final_df['commulative']=final_df['Relative'].cumsum()
# final_df['US Freq']=
us_freq=df[df['Country']=='USA']['State'].value_counts()/df[df['Country']=='USA']['State'].value_counts().sum()*100
```

```python
us_freq_frame=pd.DataFrame({'us_relative':us_freq})
us_freq_frame['commulative_freq']=us_freq_frame['us_relative'].cumsum()
us_freq_frame
```

```python
import matplotlib.pyplot as plt
import numpy as np
labels =  ['USA', 'UK', 'Belgium', 'Russia', 'Denmark', 'Germany',
       'Mexico', 'Canada']
# df = pd.DataFrame({'lab':['A', 'B', 'C'], 'val':[10, 30, 20]})
# plt.bar(x=countryfre,y=['a','b','c','d','e','f','g'])
# bp_dict = plt.bar(labels, countryfre, width=-0.4)
fig, ax = plt.subplots(figsize=(10,7))
print(fig,ax)
ax.bar(final_df.index, final_df["freq"], color="C0")
ax2 = ax.twinx()
ax2.plot( us_freq_frame["commulative_freq"], color="C1", marker="D", ms=7)
# ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")
plt.show()
```
